**Загрузим предыдущую категорию и сопоставим df_items & df_review**

```python
df_items = pd.read_csv('yamarket/small_items_21.csv')
del df_items['Unnamed: 0']
df_items.head()
```

```python
new_df_items = pd.DataFrame()
for_concat = []
for i in range(len(df_items)) :
    if (i % 200 == 0) :
        print(i)
    y = df_items.loc[i]
    if ';' in str(y['reviews']) :
        for j in y['reviews'].split(';') :
            x = df_items[df_items['prod_id'] == y['prod_id']].drop('reviews', 1)
            if (j == '') :
                continue
            x['reviews_id'] = int(j)
            #print(x.head())
            for_concat.append(x)     
```

```python
new_df_items = pd.concat(for_concat, 0)
new_df_items.head(10)
```

```python
df_reviews = pd.read_csv('yamarket/reviews.csv')
df_reviews.head()
```

```python
new_df_items.head()
```

```python
data = pd.merge(new_df_items, df_reviews, left_on = 'reviews_id', right_on = 'id', how = 'outer')
```

```python
del data['reviews_id']
del data['imgs']
data.head()
```

```python
data_without_nan = data[data['prod_id'] == data['prod_id']]
```

```python
data_without_nan.to_csv('data_{}.csv'.format(category_id))
```
